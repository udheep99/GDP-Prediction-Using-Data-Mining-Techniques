```python
data=pd.read_csv('C:/Users/hp/Downloads/countries of the world.csv',names=['Country','Region','Population','Area (sq. mi.)','Pop. Density (per sq. mi.)','Coastline (coast/area ratio)','Net migration','Infant mortality (per 1000 births)','GDP ($ per capita)','Literacy (%)','Phones (per 1000)','Arable (%)','Crops (%)','Other (%)','Climate','Birthrate','Deathrate','Agriculture','Industry','Service'])
data.head()
```

```python
data['Service'],class_names = pd.factorize(data['Service'])
```

```python
print(class_names)
print(data['Service'].unique())
```

```python
data['Country'],_ = pd.factorize(data['Country'])
data['Region'],_ = pd.factorize(data['Region'])
data['Population'],_ = pd.factorize(data['Population'])
data['Area (sq. mi.)'],_ = pd.factorize(data['Area (sq. mi.)'])
data['Pop. Density (per sq. mi.)'],_ = pd.factorize(data['Pop. Density (per sq. mi.)'])
data['Coastline (coast/area ratio)'],_ = pd.factorize(data['Coastline (coast/area ratio)'])
data['Net migration'],_ = pd.factorize(data['Net migration'])
data['Infant mortality (per 1000 births)'],_ = pd.factorize(data['Infant mortality (per 1000 births)'])
data['GDP ($ per capita)'],_ = pd.factorize(data['GDP ($ per capita)'])
data['Literacy (%)'],_ = pd.factorize(data['Literacy (%)'])
data['Phones (per 1000)'],_ = pd.factorize(data['Phones (per 1000)'])
data['Arable (%)'],_ = pd.factorize(data['Arable (%)'])
data['Crops (%)'],_ = pd.factorize(data['Crops (%)'])
data['Other (%)'],_ = pd.factorize(data['Other (%)'])
data['Climate'],_ = pd.factorize(data['Climate'])
data['Birthrate'],_ = pd.factorize(data['Birthrate'])
data['Deathrate'],_ = pd.factorize(data['Deathrate'])
data['Agriculture'],_ = pd.factorize(data['Agriculture'])
data['Industry'],_ = pd.factorize(data['Industry'])
data.head()
```

```python
data.info()
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# train the decision tree
dtree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
dtree.fit(X_train, y_train)
```

```python
# use the model to make predictions with the test data
y_pred = dtree.predict(X_test)
# how did our model perform?
count_misclassified = (y_test != y_pred).sum()
print('Misclassified samples: {}'.format(count_misclassified))
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```
